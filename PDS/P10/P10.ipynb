Pract-10: Plot the following graph for the loan datasets using Seaborn library.<br>
a) Bar Plot – Plot the number of male/female applicants.<br>
b) Strip Plot – Plot this with respect to the Applicant Income by grouping according to their Property
area.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('loan_data_set.csv')
```

```python
# a) Bar Plot – Plot the number of male/female applicants.
sns.countplot(data=df, x='Gender', palette='Set1')
plt.title('Number of Male and Female Applicants')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# b) Strip Plot – Plot this with respect to the Applicant Income by grouping according to their Property area.
sns.stripplot(data=df,x='Property_Area',y='ApplicantIncome',palette='Set1')
plt.title('Applicant Income by Property Area')
plt.xlabel('Property Area')
plt.ylabel('Income')
plt.show()
```
