Pract-3: Consider company master data of Gujarat dataset. Perform the following operations on it.<br>
a) Groupby the number of records based on the company status attribute and display their statistics.<br>
b) Groupby the number of records based on the company status and company subcategory attributes and display their statistics.

```python
import pandas as pd
```

```python
df = pd.read_csv("mca_gujarat_21042018.csv", encoding="ISO-8859-1")
```

```python
df.head()
```

```python
# a) Groupby the number of records based on the company status attribute and display their statistics.
company_status_stats = df.groupby('COMPANY_STATUS')
company_status_stats.describe()
```

```python
# b) Groupby the number of records based on the company status and company subcategory attributes and display their statistics.
company_status_and_subcategory_stats = df.groupby(['COMPANY_STATUS', 'COMPANY_SUB_CATEGORY'])
company_status_and_subcategory_stats.describe()
```
