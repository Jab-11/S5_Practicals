Pract-14: <br>Consider the loan dataset and perform standardization, normalization and encoding on it.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df = pd.read_csv('loan_data_set.csv')
df.info()
```

```python
df
```

```python
# standardization
numeric_cols=['ApplicantIncome','CoapplicantIncome','LoanAmount']
scaler = StandardScaler()
stand_df = pd.DataFrame(scaler.fit_transform(df[numeric_cols]), columns=[numeric_cols])
```

```python
s_result_df = pd.DataFrame()
for col in numeric_cols:
  s_result_df[col] = df[col]
  s_result_df[f"Standard_{col}"] = stand_df[col]
```

```python
s_result_df.head().head()
```

```python
s_result_df.describe()
```

```python
# normalization
scaler = MinMaxScaler()
norm_df = pd.DataFrame(scaler.fit_transform(df[numeric_cols]), columns=[numeric_cols])
```

```python
n_result_df = pd.DataFrame()
for col in numeric_cols:
  n_result_df[col] = df[col]
  n_result_df[f"Normalized_{col}"] = norm_df[col]
```

```python
n_result_df.head()
```

```python
n_result_df.describe()
```

```python
encoded_df = pd.get_dummies(df['Property_Area'], columns=['Property_Area'])
for index, row in encoded_df.iterrows():
    encoded_df.at[index, 'Encoded_Property_Area'] = f"{row['Rural']}{row['Semiurban']}{row['Urban']}"
encoded_df.drop(['Rural','Semiurban','Urban'],axis=1,inplace=True)
```
