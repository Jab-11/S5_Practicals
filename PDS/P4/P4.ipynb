Pract-4: Consider company master data of Gujarat dataset. Display the statistic regarding the number of records
according to the following criteria on the various attributes:<br>
a) Display the statistics of the company registered in the years 2010 to 2015.<br>
b) Display the statistics of the company whose principal business is finance and registered after 2016.<br>
c) Display the statistics of the company whose principal business is construction and authorized capital is
above 10000000.<br>
d) Display the statistics of the company according to the principal business activity and registered in the years
2015 to 2017.<br>

```python
import pandas as pd
```

```python
df = pd.read_csv("mca_gujarat_21042018.csv", encoding="ISO-8859-1")
```

```python
# Convert DATE_OF_REGISTRATION column into date format
df['DATE_OF_REGISTRATION'] = pd.to_datetime(df['DATE_OF_REGISTRATION'], format='%d-%m-%Y')
```

```python
# a) Display the statistics of the company registered in the years 2010 to 2015.
company_2010_to_2015 = df[(df['DATE_OF_REGISTRATION'].dt.year >= 2010) & (df['DATE_OF_REGISTRATION'].dt.year <= 2015)]
company_2010_to_2015.describe()
```

```python
# b) Display the statistics of the company whose principal business is finance and registered after 2016.
finance_after_2016 = df[(df['PRINCIPAL_BUSINESS_ACTIVITY_AS_PER_CIN'] == 'Finance') & (df['DATE_OF_REGISTRATION'].dt.year > 2016)]
finance_after_2016.describe()
```

```python
# c) Display the statistics of the company whose principal business is construction and authorized capital is above 10000000.
construction_cap_above_1Cr = df[(df['PRINCIPAL_BUSINESS_ACTIVITY_AS_PER_CIN'] == 'Construction') & (df['AUTHORIZED_CAP'] > 10000000)]
construction_cap_above_1Cr.describe()
```

```python
# d) Display the statistics of the company according to the principal business activity and registered in the years 2015 to 2017.
company_by_field_btw_2015_to_2017 = df[(df['DATE_OF_REGISTRATION'].dt.year >= 2015) & (df['DATE_OF_REGISTRATION'].dt.year <= 2017)].groupby('PRINCIPAL_BUSINESS_ACTIVITY_AS_PER_CIN')
company_by_field_btw_2015_to_2017.describe()
```
