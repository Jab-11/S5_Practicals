Draw graph for the given list of nodes and edges using NetworkX.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#Draw graph for the given list of nodes and edges using NetworkX.
nodes = ["A","B","C","D","E"]
edges = [("A","C"),("A","D"),("E","C"),("B","D"),("B","E")]
```

```python
G = nx.Graph()
```

```python
for n in nodes:
  G.add_node(n)
for e in edges:
  G.add_edge(e[0],e[1])
nx.draw(G,with_labels=True,node_color="red",node_size=1500,font_color="white",width=3)
plt.margins(0.2)
plt.show()
```
